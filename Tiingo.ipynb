How to get real-time data with Tiingo?

```python
data = requests.get('https://api.tiingo.com/tiingo/crypto/prices?tickers=BTCUSDT'+token).json()
data = pd.DataFrame(data[0]['priceData'])
data.set_index('date', inplace=True)
data.head()
```

How to get Forex data with Tiingo?

```python
from websocket import create_connection
import json

ws = create_connection("wss://api.tiingo.com/fx")

subscribe = {
        'eventName':'subscribe',
        'authorization':'9249c8aa0e64ad4b2a0a33d282d8df003033e6db',
        'eventData': {
            'thresholdLevel': 5
    }
}

ws.send(json.dumps(subscribe))
while True:
    print(ws.recv())
    
#ws.stop()
```
